### Pregunta 2 -  ¿Cómo se garantiza existencia de la factorización $LU$? ¿ La matriz $(\mathbf{I} - p \ \mathbf{W} \ \mathbf{D})$ está bien condicionada? ¿Cómo influye el valor de $p$?

Primero demostremos que la matriz $I-pWD$ tiene factorizacion LU. Llamemosla L por comodidad. 

Vamos a lograr la demostracion si logramos probar que la matriz $L^T$ es **EDD** (estrictamente diagonal dominante), y sus submatrices superiores izquierdas tambien lo son. Por la proposicion 4.2 del apunte, si $L$ una matriz real o compleja cuadrada EDD, entonces $L$ es inversible. Ademas, si $L$ es inversible, $L^T$ tambien lo es. 

Por definicion de EDD, planteamos
$$
\forall \; i\; , \: 1 \leqslant i \leqslant n \Rightarrow \ |(L^T)_{ii}| > \sum_{1 \leqslant j \leqslant n, j\neq i}^{n} |(L^T)_{ij}|
$$
Sabemos que las diagonales son siempre 1, o sea que es lo mismo decir 
$$
\forall \; i\;, \: 1 \leqslant i \leqslant n \Rightarrow \ 1 > \sum_{1 \leqslant j \leqslant n, j\neq i}^{n} |(L^T)_{ij}|\tag{1}
$$
Veamos si esto vale. Desarrollemos el RHS de la anterior inecuacion, 
$$
\sum_{1 \leqslant j \leqslant n,  j\neq i}^{n} |(L^T)_{ij}| = \sum_{1 \leqslant j \leqslant n, j\neq i}^{n} \left\vert-p\frac{(W^T)_{ij}}{c_i}\right\vert = \sum_{1 \leqslant j \leqslant n, 
j\neq i}^{n} p\frac{(W^T)_{ij}}{c_i}
$$

Quedaba negativo porque estos elementos se restan a $0$, y aplicando modulo quedan positivos. Como $p$ debe ser mayor que 0, y $\frac{(W^T)_{ij}}{c_i} \geq 0$, luego 

$$ \sum_{1 \leqslant j \leqslant n, j\neq i}^{n} |(L^T)_{ij}| = p * \sum_{1 \leqslant j \leqslant n, j\neq i}^{n} \frac{(W^T)_{ij}}{c_i}$$



Acotando, como p debe ser menor que 1, al multiplicar por p queda un numero menor al original. Por lo tanto,
$$
p* \sum_{1 \leqslant j \leqslant n, j\neq i}^{n} \frac{(W^T)_{ij}}{c_i} < \sum_{1 \leqslant j \leqslant n, j\neq i}^{n} \frac{(W^T)_{ij}}{c_i} = \frac{1}{c_i}* \sum_{1 \leqslant j \leqslant n, j\neq i}^{n} (W^T)_{ij}
$$
O sea que, como solo tomamos $W^T$, vale que
$$
 \; \sum_{1 \leqslant j \leqslant n,   j\neq i}^{n} (W^T)_{ij} = c_i \; 
$$
entonces vale decir
$$
\frac{1}{c_i}* \sum_{1 \leqslant j \leqslant n, j\neq i}^{n} (W^T)_{ij} = \frac{c_i}{c_i} = 1
$$
De esta forma, hemos verificado la ecuacion $(1)$.
Por lo tanto, $L^T$ es 
**EDD**




Para finalizar y verificar que $L$ tiene factorización LU,
necesitamos demostrar que si $L^T$ es EDD, entonces todas sus submatrices 
principales también son EDD.

Consideremos una submatriz principal $M$ de $L^T$ de tamaño $k \times k$.
Por definición, una submatriz principal de $L^T$ se obtiene eliminando las 
mismas filas y columnas de $L^T$. Dado que $L^T$ es EDD, para cualquier $i$ 
en $M$,
$$
|(M)_{ii}| = |(L^T)_{ii}| > \sum_{1 \leqslant j \leqslant k, j \neq i} |(M)_{ij}|
$$
Esto se debe a que eliminar filas y columnas correspondientes no 
aumenta la suma de los elementos no diagonales restantes, y la condición de EDD 
se mantiene.

En resumen, si $L^T$ es EDD, todas sus submatrices principales también son EDD, 
y por lo tanto, inversibles.

Así hemos demostrado que $L^T$ es EDD y todas sus submatrices principales 
también lo son, garantizando que $L$ tiene factorización LU.

#### Analisis de Condicion

Analizemos como influye el valor de p en el numero de condicion de matrices arbitrarias. Para ello, analizaremos dos casos. A continuacion, la implementacion de codigo para calcular y graficar el numero de condicion de la matriz L en funcion del valor de p:

```python
def calcularL(M, p): # ingresa la matriz W de conectividad
    npages = M.shape[0]
    R = matrizPuntajes(M)
    L = np.eye(npages) - p*R
    return L
```

```python
def graficarCondicion(W):
    cond=[]
    valores_p = np.linspace(0,0.9,100)

    for p in valores_p:
        cond.append(np.linalg.cond(calcularL(W,p)))

    sns.set_style("darkgrid")
    fig, ax = plt.subplots()
    ax.plot(np.arange(0,1,0.01),cond)
    plt.xlabel('Valor de p')
    plt.ylabel('Número de condición')
    plt.title("Numero de condicion en funcion de p")
    plt.show()
```

```python
W = leer_archivo(archivo_test := './tests/test_dosestrellas.txt')
graficarCondicion(W)
```

Se aprecia como para los valores entre 0 y 0.8, el numero de condicion aun se mantiene medianamente estable. Pero se ve que comienza a crecer exponencialmente. Quizas se puede ver mejor en escala logaritmica. Intentemos.

```python
def graficarCondicionLog(W):
    ps = np.logspace(0.9,0.999,100)
    cond = []
    for i in ps:
        cond.append(np.linalg.cond(calcularL(W,i)))
    
    fig, ax = plt.subplots()
    ax.plot(np.arange(0.9,1,0.001),cond)
    plt.yscale("log")
    plt.xlabel('Valor de p')
    plt.ylabel('Número de condición')
    plt.title("Numero de condicion en funcion de p")
    plt.show()
```

```python
graficarCondicionLog(W)
```

Ahora si. Podemos ver que claramente el numero de condicion crece de forma exponencial, o sea que a medida que el valor de $p \to 1$ mas nos acercamos a infinito.

Ahora vamos a ir a el test inventado por nosotros el cual tiene la estructura de un arbol binario:

(Nota: la implementacion del codigo de abajo es necesaria ya que si usabamos el codigo dado por la catedra, no se visualizaba la estructura de arbol ya que se genera casi aleatoriamente la estructura)


```python
import numpy
import networkx as nx
import matplotlib.pyplot as plt
import pydot 
import numpy as np

archivo_test = './tests/test_arbol.txt'
W = leer_archivo(archivo_test)

def dibujar_arbol(W, print_ejes=True):
    options = {
        'node_color': 'yellow',
        'node_size': 200,
        'width': 3,
        'with_labels': True
    }

    N = W.shape[0]
    G = nx.DiGraph()  # Utilizar un grafo dirigido

    # Agregar nodos al grafo
    G.add_nodes_from(range(1, N+1))

    # Agregar conexiones entre los nodos basados en la matriz de adyacencia
    for i in range(N):
        for j in range(N):
            if W[i][j] == 1:
                G.add_edge(j+1, i+1)  # Invertir la dirección del enlace

    # Generar un árbol dirigido BFS a partir del nodo raíz 1
    T = nx.bfs_tree(G, source=1)

    # Utilizar un layout más ordenado
    pos = {
        1: (0, 3),  # Nodo 1 arriba
        2: (-1, 2.5), 3: (1, 2.5),  # Nodos 2 y 3 un nivel abajo
        4: (-1.5, 2), 5: (-0.5, 2), 6: (0.5, 2), 7: (1.5, 2),  # Nodos 4, 5, 6 y 7 un nivel más abajo
        8: (-1.75, 1.5), 9: (-1.25, 1.5), 10: (-0.75, 1.5), 11: (-0.25, 1.5), 12: (0.25, 1.5), 13: (0.75, 1.5), 14: (1.25, 1.5), 15: (1.75, 1.5)  # Nodos 8 al 15 en el fondo
    }

    # Dibujar el árbol
    if print_ejes:
        print('Ejes: ', [e for e in T.edges])

    nx.draw(T, pos, **options)
    plt.show()

dibujar_arbol(W, False)
print('*'*50)
rnk, scr = calcularRanking(W,0.5)
print('Ranking:', rnk)
print('Score:', numpy.transpose(scr))
```

Podemos ver como a medida que nos desplazamos por los niveles de este arbol los score aumentan cada vez mas, es decir: $$ scr(nvl 1) < scr(nvl 2) < ... < scr(nvl Log_2(n)) < scr(nvl \ Log_2 (n + 1)) \implies max(score) = scr(nvl \ Log_2(n + 1)) $$
Por lo tanto, las paginas mejor rankeadas son las del ultimo nivel del arbol. Esto se debe a que las paginas de los primero niveles apuntan hacia los ultimos niveles, generando una acumulacion de links y dandole mas importancia a las ultimas paginas. Pero, por ejemplo, si el arbol no estuviera balanceado, es decir, si quitaramos por ejemplo las paginas 9 y 12, no valdria decir que todas las paginas del ultimo nivel tienen el mismo puntaje. Asi que lo anterior se podria profundizar mas con otro ejemplo con mas nodos. 

En conclusion, esta estructura de grafo parece interesante para una posible extension del TP, asi como estaria bueno ver que sucede con sistemas con estructura de arbol de distinto orden.

```python
def tiempoEjecucion(W, p): 
    """Calcula el tiempo de ejecucion de obtenerMaximoRankingScore en segundos

    Args:
        W: matriz de conectividad
        p: parametro del navegante aleatorio

    Returns:
        tiempo_ejecucion
    """    # calcula el tiempo de ejecucion de obtenerMaximoRankingScore 
    inicio = time.time() # inicia cronometro
    obtenerMaximoRankingScore(W, p)
    fin = time.time() # termina cronometro
    tiempo_ejecucion = fin - inicio 
    
    return tiempo_ejecucion

def tiempoEjecucionSize (dimension,p): 
    """Calcula el tiempo de ejecucion en base a la dimension de W

    Args:
        dimension: dimensión de W
        p: parametro del navegante aleatorio

    Returns:
        lista_tiempo: lista con los tiempos de ejecucion
        lista_size: lista de dimensiones paralelo a los tiempos de ejecucion
    """    
    i=2 # hago que sea minimo 2x2
    lista_tiempo=[]
    lista_size=[]
    while i <= dimension:
        W= np.random.choice([0, 1], size=(i,i)) # creo W con dim(i,i) con links random
        np.fill_diagonal(W, 0)
        tiempo_ejecucion= tiempoEjecucion(W, p)
        lista_tiempo.append(tiempo_ejecucion)
        lista_size.append(i)
        i+=1

    return lista_tiempo, lista_size

def tiempoEjecucionDensidad (densidad_maxima,p,size):
    """Calcula el tiempo de ejecucion en base a la densidad de W (densidad = links/nodos.)

    Args:
        densidad_maxima: densidad máxima de W que se grafica
        p: parametro del navegante aleatorio
        size: dimension de W (matriz cuadrada size x size)

    Returns:
        lista_tiempo: lista con los tiempos de ejecucion
        lista_densidad: lista de densidades paralelo a los tiempos de ejecucion
    """    
    lista_tiempo = []
    lista_densidad = []

    for n in range(50, densidad_maxima * 50+1,50): 
        # se elige 50 porque es un valor suficientemente alto para ver datos pero suficientemente bajo 
        #como para no romper la compu. igual cumple por lo explicado en notas. el valor de 50 es indifirente.
        W = np.zeros((size, size), dtype=int)
        np.fill_diagonal(W, 0)
        indices = np.random.choice(size * size, n, replace=False) 
        indices = np.unravel_index(indices, W.shape) # numpy magic
        W[indices] = 1
        tiempo_ejecucion= tiempoEjecucion(W, p)
        lista_tiempo.append(tiempo_ejecucion)
        lista_densidad.append(n/50)
    
    return lista_tiempo, lista_densidad
    

def regresionLin(x, y):
    # esta es una funcion auxiliar, creemos innecesario hacerle un docstring pues simplemente calcula la
    # regresion lineal dado un set de datos.
    slope, intercept, r, p, std_err = stats.linregress(x, y)   
    return np.multiply(slope, x) + intercept # es necesario pues float * lista se arregla con np.multiply :D
```

```python
def calculoProbabilidadesScore(W,subdivisiones_p):
    # la necesidad de crear subdivisiones distintas en linspace es por el rendimiento de la 
    # maquina, no tenemos un i9 de 14va generacion.
    valores_p = np.linspace(0.02, 0.99, subdivisiones_p) # nos es util para tests de 15 y 30 segs
    rnk0, scr0 = calcularRanking(W, 0.01)
    matriz_scr_p = np.array(scr0).reshape(-1, 1) 
    for p in valores_p:
        ranking, scr = calcularRanking(W, p)
        scr_columna = np.array(scr).reshape(-1, 1)
        matriz_scr_p = np.hstack((matriz_scr_p, scr_columna)) 
    return matriz_scr_p[:, 1:], valores_p 
    # Cambié matriz_scr_p[i] a matriz_scr_p[:, i] para seleccionar correctamente cada columna de matriz_scr_p.
    # Esto asegura que cada fila de matriz_scr_p tenga el mismo tamaño que valores_p. 
```

```python
def graficarProbabilidadesScore(subdiv):
    """
    Crea un gráfico de linea para visualizar la relación entre
    el valor de p y las probabilidades de las páginas.
    """
    W = leer_archivo(archivo_test)
    npages = W.shape[0]
    matriz_scr_p, valores_p = calculoProbabilidadesScore(W,subdiv)
    
    sns.set_style(style="darkgrid") # fashion
    plt.figure(figsize=(8, 6)) # tamaño
    for i in range(npages):
        plt.plot(valores_p, matriz_scr_p[i])
    plt.xlabel('Valor de p')
    plt.ylabel('Puntaje de pagina')
    plt.title('Puntaje en funcion de parametro p')
    plt.show()
```
